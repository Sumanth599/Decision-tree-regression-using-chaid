```python
class DecisionTreeRegressor:
    def __init__(self, max_depth=None, min_samples_split=2):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        
    def fit(self, X, y):
        self.X = X
        self.y = y
        self.tree = self._grow_tree()
        
    def predict(self, X):
        return [self._predict(inputs) for inputs in X]
        
    def _grow_tree(self,tree, depth=0):
        num_samples, num_features = self.X.shape
        if depth == self.max_depth or num_samples < self.min_samples_split:
            return {'leaf': True, 'value': np.mean(self.y)}
        
        feature_indices = np.random.choice(num_features, np.sqrt(num_features), replace=False)
        
        best_feature, best_threshold = self._best_criteria(feature_indices)
        left_indices, right_indices = self._split(self.X[:, best_feature], best_threshold)
        
        left_subtree = self._grow_tree(depth+1)
        right_subtree = self._grow_tree(depth+1)
        
        return {'feature': best_feature,
                'threshold': best_threshold,
                'left': left_subtree,
                'right': right_subtree}
    
    def _best_criteria(self, feature_indices):
        best_reduction = -np.inf
        split_index, split_threshold = None, None
        for feature_index in feature_indices:
            X_column = self.X[:, feature_index]
            thresholds = np.unique(X_column)
            for threshold in thresholds:
                reduction = self._standard_deviation_reduction(X_column, threshold)
                if reduction > best_reduction:
                    best_reduction = reduction
                    split_index = feature_index
                    split_threshold = threshold
        return split_index, split_threshold
    
    def _standard_deviation_reduction(self, X_column, split_threshold):
        parent_std = np.std(self.y)
        
        left_indices, right_indices = self._split(X_column, split_threshold)
        if len(left_indices) == 0 or len(right_indices) == 0:
            return 0
        
        left_std = np.std(self.y[left_indices])
        right_std = np.std(self.y[right_indices])
        
        child_std = (len(left_indices) / len(self.y)) * left_std + \
                    (len(right_indices) / len(self.y)) * right_std
        
        return parent_std - child_std
    
    def _split(self, X_column, split_threshold):
        left_indices = np.argwhere(X_column <= split_threshold).flatten()
        right_indices = np.argwhere(X_column > split_threshold).flatten()
        return left_indices, right_indices
    
    def _predict(self, inputs):
        tree = self.tree
        while not tree['leaf']:
            if inputs[tree['feature']] <= tree['threshold']:
                tree = tree['left']
            else:
                tree = tree['right']
        return tree['value']
```
